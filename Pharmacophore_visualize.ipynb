```python
#This is prepared to run in environment "chem_viz"
import xml.etree.ElementTree as ET
import py3Dmol
```

```python
def visualize_stabilized_pml(pml_path, protein_path=None):
    # Setup high-res view for your workstation
    view = py3Dmol.view(width=1200, height=800)
    
    if protein_path:
        with open(protein_path, 'r') as f:
            view.addModel(f.read(), 'pdb')
        view.setStyle({'cartoon': {'color': 'lightgray', 'opacity': 0.6}})
    
    tree = ET.parse(pml_path)
    root = tree.getroot()
    
    # Official LigandScout Color Palette
    color_map = {
        "H": "#f1c40f",           # Hydrophobic (Yellow)
        "HBA": "#e74c3c",         # H-Bond Acceptor (Red)
        "HBD": "#3498db",         # H-Bond Donor (Blue - Note: standard LS is often blue/white)
        "AR": "#e67e22",          # Aromatic Ring (Orange)
        "PI": "#9b59b6",          # Positive Ionizable (Purple)
        "NI": "#c0392b",          # Negative Ionizable (Dark Red)
        "exclusion": "#2c3e50"    # Exclusion Volume (Dark Gray/Black)
    }

    # Helper to safely extract coordinates
    def get_coords(elem):
        if elem is None: return None
        return {'x': float(elem.get('x3')), 'y': float(elem.get('y3')), 'z': float(elem.get('z3'))}

    # 1. PROCESS POINTS (H, AR, PI, NI)
    for pt in root.findall(".//point"):
        # Prioritize 'name' for chemical features
        feat_type = pt.get('name') or pt.get('type')
        print(feat_type)
        pos = get_coords(pt.find("position"))
        if pos:
            view.addSphere({
                'center': pos,
                'radius': float(pt.find("position").get('tolerance', 1.0)),
                'color': color_map.get(feat_type, "magenta"),
                'opacity': 0.8
            })

    # 2. PROCESS VECTORS (HBA, HBD)
    for vec in root.findall(".//vector"):
        feat_type = vec.get('name') or vec.get('type')
        print(feat_type)
        origin = get_coords(vec.find("origin"))
        target = get_coords(vec.find("target"))
        
        if origin and target:
            # Main sphere
            view.addSphere({
                'center': origin, 
                'radius': 0.7, 
                'color': color_map.get(feat_type, "magenta"), 
                'opacity': 0.9
            })
            # Interaction direction
            view.addArrow({
                'start': origin, 'end': target,
                'radius': 0.15, 'color': color_map.get(feat_type, "magenta"),
                'opacity': 0.8
            })

    # 3. PROCESS VOLUMES (Strictly Exclusion)
    for vol in root.findall(".//volume"):
        # If it's a <volume> tag, in your workflow it's almost certainly an exclusion
        # We ignore any 'type' that claims to be HBD inside a volume tag to avoid the bug
        pos = get_coords(vol.find("position"))
        if pos:
            view.addSphere({
                'center': pos,
                'radius': float(vol.find("position").get('tolerance', 1.0)),
                'color': color_map["exclusion"],
                'opacity': 0.25  # Low opacity for exclusions is standard
            })

    view.zoomTo()
    return view.show()
```

```python
visualize_stabilized_pml("../Pharmacophores/pharmacophore_west_to_center-back_east.pml", "../receptor_only.pdb")
```
